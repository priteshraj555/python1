```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
from sklearn.cluster import  KMeans
```

```python
import seaborn as sns
```

```python
import os
```

```python
#read data
folder_path=r"C:\Users\USER\Desktop\pythooon api\downloads"
```

```python
image_file=os.listdir(folder_path)
```

```python
image_file[0:5]
```

```python
#image_file_path
image_file_path=[os.path.join(folder_path,x) for x in image_file]
image_file_path[0:5]
```

```python
# Get Images Data
def image_process(image):
    df=pd.DataFrame()
    img=cv2.imread(image)
    #  resize
    img_res=cv2.resize(img,(80,80))
    img_ret=np.array(img_res)
    img_flat=img_ret.flatten()
    df=pd.DataFrame(img_flat).T
    return df    
```

```python
df=image_process(r'C:\Users\USER\Desktop\pythooon api\downloads\cat_10.jpg')
df
```

```python
# process image
dfx=pd.DataFrame()
for image in image_file_path:
    dfr=image_process(image)
    dfx=pd.concat([dfr,dfx])
```

```python
dfx
```

```python
dfx.to_csv('df_csv.csv',index=True)
dfx.head()
```

# Implementing PCA

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca_df=pca.fit_transform(dfx)
```

```python
pca_df=pd.DataFrame(pca_df)
```

```python
plt.title("Data Before Clustering")
sns.scatterplot(x=0, y=1, data=pca_df)
```

# Implement KMeans Clustering

```python
wcss=[]
for x in range (2,10):
    kmeans=KMeans(n_clusters=x)
    kmeans.fit_predict(dfx)
    wcss.append(kmeans.inertia_)
```

```python
#plot the inertia value
plt.plot(range(2,10), wcss, marker='D')
plt.grid()
```
