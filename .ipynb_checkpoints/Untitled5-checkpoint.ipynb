```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
data=pd.read_csv(r'cars_ready_to_eda.csv')
```

```python
data['Fuel_Type'].nunique()
```

```python
data['Fuel_Type'].unique()
```

```python
data.shape
```

```python
sns.countplot(x='Fuel_Type',data=data,palette='cool')
```

```python
sns.barplot(x='Fuel_Type',y='Mileage',data=data,palette='cool')
```

```python
#calculate the average of power with respect to location
plt.title("location and power Distribution")
sns.barplot(x='Location',y='Power',data=data,palette='cool')
plt.grid()
plt.xticks(rotation=90)
```

```python
#count the numbr of
data
```

```python
sns.lineplot(x='Location',y='Mileage',ci=None,data=data)
plt.xticks(rotation=90)
```

```python
sns.lineplot(x='Location',y='Mileage', marker='D',ci=None,data=data)
plt.xticks(rotation=90)
```

```python
sns.lineplot(x='Location',y='Mileage', marker='D',ci=None,data=data,label='Avg Mileage')
sns.lineplot(x='Location',y='Power',data=data,label='avg power')
plt.xticks(rotation=90)
```

```python
sns.scatterplot(x='Power',y='Mileage', hue='Fuel_Type',data=data)
```

# Strip plot

```python
sns.stripplot(x='Power',y='Mileage', hue='Fuel_Type',data=data)
```

```python
sns.stripplot(x='Fuel_Type',y='Mileage', hue='Owner_Type',data=data)
```

# box plot

```python
sns.boxplot(x='Mileage',data=data)
```

```python
sns.boxplot(y='Mileage',data=data)
```

```python
sns.boxplot(y='Power',hue='Fuel_Type',data=data)
```

```python
sns.kdeplot(x='Power',hue='Fuel_Type',data=data)
```

```python
sns.histplot(x='brand',kde=True,hue='Fuel_Type',data=data)
plt.xticks(rotation=90)
```

# Pei chart

```python
data['Fuel_Type'].value_counts()
```

```python
plt.pie(data['Fuel_Type'].value_counts(),labels=data['Fuel_Type'].value_counts().index,autopct='%1.1f%%',startangle=90)
```

```python
plt.pie(data['Owner_Type'].value_counts(),labels=data['Owner_Type'].value_counts().index,autopct='%1.1f%%',startangle=90)
plt.xticks(rotation=90)
```
